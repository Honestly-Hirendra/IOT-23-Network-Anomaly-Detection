```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.metrics import classification_report

# Load preprocessed data
data = pd.read_csv('D:\Important\Projects\IOT_23\Models\iot23_Preprocessed.csv')

# Select features for DBSCAN (e.g., numeric features)
X = data.drop(['label' , 'encoded_label'], axis=1)  # Ensure your data has a 'label' column for evaluation

# Fit DBSCAN model
dbscan = DBSCAN(eps=0.10, min_samples=5).fit(X)

# Add labels to data
data['dbscan_labels'] = dbscan.labels_
data['dbscan_binary'] = (data['dbscan_labels'] == -1).astype(int)

# Evaluation using True Data
y_true = (data['encoded_label'] == -1).astype(int)  # malicious = 1
y_pred = data['dbscan_binary']

# Generate evaluation metrics
print(classification_report(y_true, y_pred, target_names=["Benign", "Malicious"]))
```
